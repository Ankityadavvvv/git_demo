**Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
def categorize_size(bedrooms):
    if bedrooms <= 10:
        return 'small'
    elif bedrooms < 1000:
        return 'medium'
    else:
        return 'large'


housing['total_bedroom_size'] = housing['total_bedrooms'].apply(categorize_size)
housing.total_bedroom_size
```

This code first defines a function called categorize_size() that takes a single argument (bedrooms) and returns a string indicating the size category based on the value of bedrooms. If bedrooms is less than or equal to 10, the function returns 'small'. If bedrooms is greater than 10 but less than 1000, the function returns 'medium'. If bedrooms is 1000 or greater, the function returns 'large'.apply() function, and assigns the results to a new column called total_bedroom_size.
